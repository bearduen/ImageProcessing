```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
#建立power-law查找表
def creat_lookup_table(gamma):
    table = np.array([(i / 255.0) ** gamma * 255.0 for i in range(0, 256)]).astype("uint8")
    return table
```

```python
#影像處理
def adjust(gamma, img_name):
    table = creat_lookup_table(gamma)
    
    #讀入影像
    img = cv.imread(img_name, cv.IMREAD_COLOR)

    #將影像mapping
    img_row = img.shape[0]
    img_col = img.shape[1]
    for i in range(0, img_row):
        for j in range(0, img_col):
            for k in range(0, 3):
                img[i][j][k] = table[img[i][j][k]]
            
    #將處理過的影像存回電腦
    cv.imwrite("gamma="+str(gamma)+"_"+img_name, img)
    
    #顯示影像
    img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
    plt.imshow(img)
    plt.axis('off')
    plt.show()
```

```python
for gamma in [0.2, 0.5, 1.5, 5.0]:
    #gamma=0.2 0.5 1.5 5.0
    adjust(gamma, "01.jpg")
```
